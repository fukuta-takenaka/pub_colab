```python
# GitHubからcsv読み込み、死者数上位の表示

import pandas as pd
df_deaths = pd.read_csv('https://raw.githubusercontent.com/fukuta-takenaka/pub_colab/main/data/2019_%E6%84%9F%E6%9F%93%E7%97%87_%E6%AD%BB%E8%80%85%E6%95%B0.csv')
df_deaths.sort_values('総数', ascending=False).head(20)
```

```python
# 検査数データの読み込み
df_tests = pd.read_csv('https://github.com/fukuta-takenaka/pub_colab/raw/main/data/2019_%E6%A4%9C%E6%9F%BB%E7%B7%8F%E6%95%B0_%E5%A4%96%E6%9D%A5.csv')
df_tests.head()
```

```python
# 例ではマイコプラズマについて図にします
# 死亡者データからマイコプラズマのレコードを抽出
myco_death_records = df_deaths.query('病名.str.startswith("マイコプラズマ")', engine='python')
myco_death_records
```

```python
# 検査数データからマイコプラズマのレコードを抽出
myco_test_records = df_tests.query('診療行為.str.contains("マイコプラズマ")', engine='python')
myco_test_records.sort_values('総計', ascending=False)
```

```python
# 最も検査数の多い"マイコプラズマ抗体定性"と死亡数の月別推移を図示します
# 死亡者数は2019年1月～12月なのに対し、検査数は2019年4月～2020年3月になっているため、オーバーラップしている部分のみを使用します
myco_deaths = myco_death_records.iloc[0, 6:]
print('マイコプラズマ死者数\n', myco_deaths)
myco_tests = myco_test_records.query('診療行為 == "マイコプラズマ抗体定性"').iloc[0, 6:15]
print('マイコプラズマ検査数\n', myco_tests)

# データの抽出・加工はここまでで完了です
```

```python
# matplotlibで可視化します
# コード量が多く見えますが、軸ラベルの設定などがほとんどなので、最低限必要なコード（*）は少ないです
# どの行をコメントアウトするとグラフがどう変わるか確認すると、コードの意味がわかりやすいです
!pip install japanize-matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
japanize_matplotlib.japanize()

sns.set(font="IPAexGothic")
fig, ax1 = plt.subplots()  # (*)
ax1.plot(myco_tests, label='検査数（左軸）')  # (*)
ax1.set_ylim(0, 50000)
ax1.set_ylabel('検査数')
ax1.set_xlabel('2019年')

ax2 = ax1.twinx()  # (*) 右軸の生成
ax2.bar(range(len(myco_deaths)), myco_deaths, alpha=0.4, color='green', label='死亡数（右軸）')  (*)
ax2.set_ylim(0, 10)
ax2.set_ylabel('死亡数')

# 凡例
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2, loc=2, borderaxespad=0.)

plt.savefig('mycoplasma.png', dpi=300, bbox_inches='tight')  (*)
plt.show()  (*)
```

患者数は秋～冬にかけて増加する傾向が見られ、1年間で3～4倍程度の増減がありそうです。一方で、2019年の検査数は年間を通して比較的安定しており、否定試験としての利用が多いのではないかと思われます。

```python
# 参考 GitHubのcsv生成
from io import BytesIO
import openpyxl
import requests

# （下準備：GitHubファイルの生成）検査総数、診療月別算定回数 (2019年)（NDB) エクセルからcsvに変換してドライブに保存
url = 'https://www.mhlw.go.jp/content/12400000/000825122.xlsx'
r = requests.get(url, allow_redirects=True)
wb = openpyxl.open(BytesIO(r.content))

for sheet_name in ['外来', '入院']:
  ws = wb[sheet_name]
  cols = [cell.value.replace('\n', '') for cell in ws['A3:R3'][0]]
  data_cells = ws[f'A5:R{ws.max_row}']

  # 分類コード・分類名称が先頭行以外省略されているので補完
  data = list()
  cls_code = ''
  cls_name = ''
  for row in data_cells:
    if row[0].value:
      cls_code = row[0].value
      cls_name = row[1].value
    data.append([cls_code, cls_name] + [cell.value for cell in row[2:]])

  df = pd.DataFrame(data, columns=cols)
  # 10件未満をハイフンとしているが、簡便さのためゼロで置換
  df = df.replace('-', 0)
  df.to_csv(f'2019_検査総数_{sheet_name}.csv', index=False)


  # （下準備：GitHubファイルの生成）感染症による死亡数（月別. 2019年）（e-Stat）csvを整形して保存
# csvファイルだが列数がそろっていないので整形が必要
url = 'https://www.e-stat.go.jp/stat-search/file-download?statInfId=000031982915&fileKind=1'
content = requests.get(url, allow_redirects=True).content.decode('shift_jis')
lines = content.split('\r\n')

# 5行目がカラム名, 分類コード+病名は分割してカラム名追加
cols = ['分類コード', '病名'] + lines[4].split(',')[1:]

# 8行目から119行目までがデータ範囲
data = lines[7:119]

# 分類コードと病名を分割
data = [x.replace(' ', ',', 1) for x in data]

data = [x.split(',') for x in data]
df = pd.DataFrame(data, columns=cols)
# 10件未満をハイフンとしているが、簡便さのためゼロで置換
df = df.replace('-', 0)
df.to_csv('2019_感染症_死者数.csv', index=False)
```
